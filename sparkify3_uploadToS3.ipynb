Import necessary libraries

```python
import boto3
import os
```

Connect to S3 bucket after configuring AWS via CLI

```python
s3_client = boto3.client('s3') # low level client interface
buckets = s3_client.list_buckets()

s3_resource = boto3.resource('s3') # high level interface

#print existing S3 buckets
for bucket in buckets['Buckets']:
    print(bucket['CreationDate'].ctime(), bucket['Name'])
```

Create a bucket

```python
# bucket name must be unique throughout the enitre AWS platform

session = boto3.session.Session()
current_region = session.region_name
bucket_name = 'sparkify3-logs'
bucket_response = s3_client.create_bucket(Bucket=bucket_name)
print(bucket_name, current_region)
```

```python
for file in file_names:
    s3_client.upload_file(path + file, bucket_name, file) # the third argument is the key
        
# len(obj)
```
